```python
import numpy as np
from sklearn import datasets, linear_model, metrics
```

```python
## Load the diabetes dataset
diabetes = datasets.load_diabetes()
diabetes_X = diabetes.data # matrix of dimensions 442x10

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
```

```python
X = diabetes_X_train # (422x10)
y = diabetes_y_train # (422,) it is a vector not a matrix 
m,n=X.shape # shape is (422,10)

#adding the X0 to the X matrix
X0=np.ones((m,1))
X=np.hstack((X0,X))
m,n=X.shape # (422,11)
y=np.reshape(y,(m,1)) # (422x1) now a matrix

#initialising thetas with zeros
Theta=np.zeros((n,1)) # (nx1)

#taking transpose to get Theta as (1Xn) and X as (nXm)
H_0x=np.dot(Theta.T,X.T) # (1xm)
H_0x=H_0x.T # (mx1)

#calculating cost function
diff=H_0x-y # (mx1)
sum1=np.dot(diff.T,diff) # (1x1)
J0=sum1/(2*m) 
J0 #single value

epochs=100000 #number of iterations
alpha=0.5 #learning rate

fin_Theta=np.zeros(n) # (1xn)
fin_Theta=fin_Theta.T # (nx1)
#calculate the gradients
for i in range (epochs):
    change=np.dot((H_0x-y).T,X) 
    change=(change*alpha)/m
    Theta=Theta-change.T #updating theta values
    
    # again calculating the error 
    H_0x=np.dot(Theta.T,X.T)
    H_0x=H_0x.T
    diff=H_0x-y
    sum1=np.dot(diff.T,diff)
    # if error is less update Theta array
    if i%5000 ==0 :
        print ("%d: %f" % (i, J0))
    if J0>sum1/(2*m):
        J0=sum1/(2*m)
        fin_Theta=Theta
print ("The Theta values are: \n")
print (fin_Theta)
```

```python
#Test
X_test=diabetes_X_test
y_test=diabetes_y_test

#adding column of ones as X0
m,n=X_test.shape
y_test=np.reshape(y_test,(m,1))
X0=np.ones((m,1))
X_test=np.hstack((X0,X_test))

#hypothesis
H_0x=np.dot(fin_Theta.T,X_test.T)
H_0x=H_0x.T

for i in range (len(H_0x)):
    print (H_0x[i],"-> ",y_test[i])
print('Coefficients: \n', fin_Theta)
print("Mean squared error: %.2f" % J0)
print("="*120)
```
